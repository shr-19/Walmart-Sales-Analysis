```python
Country_sales=df.Sales.mean()
Country_profit=df.Profit.mean()
Country_sales,Country_profit
```

```python
Country_pie=pd.DataFrame({'Attributes':["Country_sales","Country_profit"],
                        'Mean':[226.4932656884171,33.849138307836405]})
Country_pie.set_index('Attributes', inplace=True)
pl = Country_pie.plot.pie(subplots=True, title="Walmart Sales Analysis in United States", legend=False,
                   autopct='%1.1f%%', startangle=0,figsize=(5,5))
```
